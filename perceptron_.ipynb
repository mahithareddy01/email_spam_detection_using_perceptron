## 🏢Business Scenario: **Email Spam Detection for a Startup**
A startup company is building a basic email filtering system to classify emails as:<br>
**1** → Spam<br>
**0** → Not Spam<br>
The company wants to first implement a simple Perceptron model before moving to complex neural networks.<br>
They identified **two key binary features** from emails:<br>
- **Contains Suspicious Keywords** (e.g., “lottery”, “prize”, “urgent”)<br>
    Yes → 1<br>
    No → 0<br>
- **Sender is Unknown**<br>
    Yes → 1<br>
    No → 0<br>
So each email is represented as:<br>
X = (x1,x2)<br>
Where:<br>
 - x1 = Suspicious keywords<br>
 - x2 = Unknown sender

```python
import numpy as np
```

```python
X = np.array([[1,1],[1,0],[0,1],[0,0]])
y = np.array([1,1,1,0])
```

### **Part 1 – Conceptual Understanding**
- Is this dataset linearly separable?
- Can a single-layer Perceptron solve this problem?
- What logical function does this resemble?

- The dataset is linearly separable.
- single layer perceptron can solve this problem.
- It resembles OR function.

### **Part 2 - Mathematical Formulation**

```python
w = np.zeros(2)
b = 0
learning_rate = 0.1
epochs = 5
print(f"Initial weghts :{w}")
print(f"Initial bias:{b}")
```

```python
def step_function(z):
    return 1 if z>=0 else 0
```

```python
for epoch in range(epochs):
    print(f"Epoch {epoch}")
    for i in range(len(X)):
        z = np.dot(w,X[i])+b
        y_pred = step_function(z)
        err = y[i]-y_pred
        w = w+learning_rate*err*X[i]
        b = b+learning_rate*err
        print(f"Input:{X[i]},Actual:{y[i]},prediction:{y_pred},w:{w},b:{b}")
```

```python
print("Final Weight:",w)
print("Final Bias:",b)
```

### **Part 3 - Decision Boundary**
After training:
1. Write the final equation of the decision
boundary.
2. Plot the decision boundary.

```python
import matplotlib.pyplot as plt
for i in range(len(X)):
    if(y[i]==0):
        plt.scatter(X[i][0],X[i][1],marker='x')
    else:
        plt.scatter(X[i][0],X[i][1],marker='o')
x_vals = np.array([-0.2,1.2])
y_vals = -(w[0]*x_vals+b)/w[1]
plt.plot(x_vals,y_vals)
plt.title('Decision Boundary')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
